```python
import sympy as sp
```

```python
x = sp.Symbol('x')
expr = sp.sin(x) ** 2 - sp.cos(x) ** 2
expr
```

```python
# Корни функции
expr_solve = sp.solveset(expr, x, domain = sp.Reals)
expr_solve
```

```python
# Нахождение производной
expr_diff = sp.diff(expr)
expr_diff
```

```python
# Построение графика
graph = sp.plot(expr)
```

```python
# Возрастание функции
rise_exp = sp.solveset(expr_diff > 0, x, domain = sp.Reals)
rise_exp
```

```python
# Убывание функции
fall_exp = sp.solveset(expr_diff < 0, x, domain = sp.Reals)
fall_exp
```

```python
# Экстремумы
peak_exp = sp.solveset(expr_diff, x, domain = sp.Reals)
peak_exp
```

```python
# Функция больше нуля (f(x) > 0)
peak_exp_origin = sp.solveset(expr > 0, x, domain = sp.Reals)
peak_exp_origin
```

```python
# Функция меньше нуля (f(x) < 0)
fall_exp_origin = sp.solveset(expr < 0, x, domain = sp.Reals)
fall_exp_origin
```
